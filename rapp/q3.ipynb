```python
import pandas as pd
import numpy as np
```

```python
def flatten_columns(df):
    df.columns = ['_'.join(col[::-1]).strip() for col in df.columns.values]
    return df

def summarize(df: pd.DataFrame, col_names: list, group_by_var='a categorical variable', funs=['sum', 'mean']) -> pd.DataFrame:
    res = df[col_names + [group_by_var]].groupby(group_by_var).aggregate(funs)
    return flatten_columns(res).reset_index()
```

```python
df = pd.read_csv('hsbdemo.csv')
print(list(df))
df.head()
```

```python
res = summarize(df, ['read', 'write', 'math', 'science'], group_by_var='prog', funs=['sum', 'mean', 'min', 'max'])
res
```

```python
flatten_columns(res).reset_index()
```
